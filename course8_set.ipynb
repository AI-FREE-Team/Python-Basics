# Set ( 元組 )

* set() 函數創建一個無序不重複元素集，可進行關係測試，刪除重複數據，還可以計算交集、差集、並集等。

```
# class set([iterable])
# iterable -- 可迭代對象對象；
```

## 實例

* 以下實例展示了 set 的使用方法：

```
x = set('runoob')
y = set('google')
print(x,y)

x, y
(set(['b','r','u','o','n']), set(['e','o','g','l'])) # 重複的被刪除
print(x,y)

x & y  # 交集
set(['o'])
print(x,y)

x | y  # 並集
set(['b', 'e', 'g', 'l', 'o', 'n', 'r', 'u'])
print(x,y)

x - y  # 差集
set(['r','b','u','n'])
print(x,y)
```
