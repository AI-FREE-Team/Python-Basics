# 運算子(Operator)

* 這個章節，我們將教導什麼事運算子，並用一些實例來引導大家了解這個單元

* 基本運算

```
a=1
b=2
c=a+b
d=a-b
e=a*b
```

* 並透過print()來檢查

```
print("a",a)
print("b",b)
print("c",c)
print("d",d)
print("e",e)
```

* 除法需要特別注意，/為取商，%為取餘數。

```
a=2/3
b=2%3
print("a",a)
print("b",b)
```

* 邏輯運算，回傳為 True 或 False。

```
a==b #a是否等於b
a>b  #a是否大於b
a<b  #a是否小於b
a>=b  #a是否大於等於b
a<=b  #a是否小於等於b
a!=b  #a是否不等於b
```
