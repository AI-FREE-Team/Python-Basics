## While 迴圈

* Python 編程中 while 語句用於循環執行程序，即在某條件下，循環執行某段程序，以處理需要重複處理的相同任務。其基本形式爲：

```
# while(繼續的條件)：
  # 當條件成立時,會執行的地方
```

* 執行語句可以是單個語句或語句塊。判斷條件可以是任何表達式，任何非零、或非空（null）的值均爲true。

* 當判斷條件假false時，迴圈中止。

* 執行流程圖如下：

![流程圖](https://d33wubrfki0l68.cloudfront.net/70ac2262ad7f6cd301ba93829e8dbe5c2ee81f48/44d6c/wp-content/uploads/2017/06/while-loop-1.png)


* 是不是很簡單呢,馬上看實例

```
c = 0
while (c < 9):
   print('The count is:', c)
   c = c + 1

print('')
print('Good bye!')
print('c =',c)
```

* while 語句時還有另外兩個重要的命令 **continue，break** 來跳過循環，continue 用於**跳過該次循環**，break 則是用於**退出循環**，此外"判斷條件"還可以是個**常值**，表示循環必定成立，具體用法如下：

```
i = 1
while i < 10:   
    i += 1
    if(i%2)> 0:     # 非雙數時跳過輸出
        continue
    print('pass!!')
    print(i)         # 輸出雙數2、4、6、8、10
 
j = 1
while 1:            # 循環條件爲1必定成立
    print(j)         # 輸出1~10
    j += 1
    if j > 5:     # 當i大於10時跳出循環
        break
print('end')
```

## 無限循環


* 如果條件判斷語句**永遠爲 true**，循環將會**無限**的執行下去，如下實例：

```
inifinty = 1
num = inifinty
while inifinty == 1 :  # 該條件永遠爲true，循環將無限執行下去
   num = num + 1
   print("You entered: ",num)
 
print("Never!")
# 記得手動停止
```

## 迴圈使用else語句

* 在 python 中，while … else 在迴圈條件爲 false 時執行 else 語句塊：

```
count = 0
while count < 5:
   print(count, "is < than 5")
   count = count + 1
else:
   print(count, "is = = than 5")
```
