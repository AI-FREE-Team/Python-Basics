# Ranges ( 範圍 )




*   Range 內的元素是不可變的，並且可以使用索引，通常用於迴圈(Loop)
*   宣告方式為 range( start , stop, step )
*   start : 計數從 start 為起點。如果沒有數字就是從0開始。
*   stop : 計數到 stop 為終點。但不包括 stop 的數字。
*   step : 移動步長，沒有數字就視為 1 。


```
for i in range(5):
  print(i,end=",")
print()
print("0 ~ 4,但不包括5")

print()
for i in range(3,10):
  print(i,end=",")
print()  
print("從3開始,但不包括10")

print()
for i in range(3,10,2):
  print(i,end=",")
print()  
print("從3開始,但不包括10,每次跳2")

print()
print("可以遞增也可以遞減")
for i in range(10,-30,-5):
  print(i,end=",")
print()  
print("從10開始,但不包括-30,每次跳-5")
```

*  來看看其他用法

```
words = ['jane','john','mark','harry','mike','ed']
wordlist = []
for i in range(len(words)):
  wordlist.append([i,words[i]])
  print(wordlist)
```

*  也可以把 range 宣告成變數

```
r = range(0,10,2)
print(r)
print(list(r))
print(2 in r)
print(r.index(2))
print(r[3])
print(r[:3])
print(r[-1])
```
